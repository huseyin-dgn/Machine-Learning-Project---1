```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2 , include_bias=True)

poly_feat = poly.fit_transform(x)
```

# TRAN TEST SPLT

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    poly_feat, y, test_size=0.33, random_state=42)
```

# LNEER REGR

```python
from sklearn.linear_model import LinearRegression
```

```python
lineer = LinearRegression(fit_intercept=True)
```

```python
lineer.fit(X_train,y_train)
```

```python
lin_pred = lineer.predict(X_test)
```

# DEERLENDRME

```python
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
```

```python
mae = mean_absolute_error(y_test , lin_pred)
rmse = np.sqrt(mean_squared_error(y_test , lin_pred))
r2 = r2_score(y_test , lin_pred)

print("Mae :", mae)
print("Rmse :", rmse)
print("Rr2  :",r2 )
```

# EN Y DEER

```python
train_rmse = []
test_rmse = []

for d in range(1,6):
    
    poly= PolynomialFeatures(degree=d , include_bias=False)
    # Veri trn azaltmak

    poly_feat = poly.fit_transform(x)

    X_train, X_test, y_train, y_test = train_test_split(
    poly_feat, y, test_size=0.33, random_state=42)

    lineer = LinearRegression(fit_intercept=True)
    lineer.fit(X_train ,y_train)
    train_pred = lineer.predict(X_train)
    test_pred = lineer.predict(X_test)
    
    # Eitim ve test setleri iin RMSE hesapla
    train_RMSE = np.sqrt(mean_squared_error(y_train, train_pred))
    test_RMSE = np.sqrt(mean_squared_error(y_test, test_pred))

    # Hesaplanan RMSE deerlerini listelere ekle
    train_rmse.append(train_RMSE)
    test_rmse.append(test_RMSE)

    
# BELLEK KULLANIMINIAZ BALI GEREKLEECEKTR
#1. Polinom Derecesinin Artrlmas:
#Kodda PolynomialFeatures(degree=d) kullanlarak polinom derecesi artrlyor. 
# Ancak, ok yksek dereceler bellei ar ekilde kullanabilir. 
# Eer bellek hatas alyorsanz, dereceyi 3'ten fazla yapmamaya zen gsterin.
```

```python
plt.plot(range(1,8),train_rmse[:7],label='TRAIN')
plt.plot(range(1,8),test_rmse[:7],label='TEST')
plt.xlabel("Polinom Derecesi")
plt.ylabel("RMSE")
plt.legend()
```

# FNAL

```python
# POLY REGR
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3 , include_bias=True)

poly_feat = poly.fit_transform(x)


# TRAN TEST SPLT
x = df.drop("CO2_Emissions" , axis=1)
y=df["CO2_Emissions"]
from sklearn.preprocessing import LabelEncoder

# Label encoding 
# Saysal veri yap
le = LabelEncoder()

categorical_columns = x.select_dtypes(include=['object']).columns

for column in categorical_columns:
    x[column] = le.fit_transform(x[column])
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    poly_feat, y, test_size=0.33, random_state=42)

# LNEER REGR

from sklearn.linear_model import LinearRegression
lineer = LinearRegression(fit_intercept=True)
lineer.fit(X_train,y_train)
lin_pred = lineer.predict(X_test)

# DEERLENDRME
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
mae = mean_absolute_error(y_test , lin_pred)
rmse = np.sqrt(mean_squared_error(y_test , lin_pred))
r2 = r2_score(y_test , lin_pred)

print("Mae :", mae)
print("Rmse :", rmse)
print("Rr2  :",r2 )
```

```python
df["CO2_Emissions"].describe()
```

```python
# Gerek ve tahmin edilen deerleri karlatrmak iin scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, lin_pred, color='blue', alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], '--r', label='Perfect Prediction')  # Yksek doruluk izgisi
plt.xlabel("Gerek CO2 Emisyonlar")
plt.ylabel("Tahmin Edilen CO2 Emisyonlar")
plt.title("Gerek vs Tahmin Edilen CO2 Emisyonlar")
plt.legend()

```

```python
# Model performans metriklerini bar grafikte gster
metrics = {'MAE': mae, 'RMSE': rmse, 'R': r2}
plt.figure(figsize=(8, 5))
plt.bar(metrics.keys(), metrics.values(), color=['blue', 'orange', 'green'])
plt.title("Model Deerlendirme Metrikleri")
plt.xlabel("Metrikler")
plt.ylabel("Deerler")

```

```python
# Modelin hata dalmn incelemek
residuals = y_test - lin_pred

plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True, color='green', bins=30)
plt.title("Model Residuals Distribution")
plt.xlabel("Residuals (Hatalar)")
plt.ylabel("Frekans")

```
